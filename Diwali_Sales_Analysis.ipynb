# Exploratory Data analysis 

Gender

```python
df.columns
```

```python
# this is only Count for Gender
ax = sns.countplot(x='Gender',data= df)

# below code to count and displayed also bar label 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
Sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = Sales_gen,)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

Age

```python
df.columns
```

```python
ax= sns.countplot(x='Age Group',hue='Gender',data=df)

for datas in ax.containers:
    ax.bar_label(datas)
```

```python
# total Amount vs Age Group
sales_Age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Age Group' , y = 'Amount' ,data =sales_Age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

State

```python
df.columns
```

```python
# total amount/sales from top 10 states
slaes_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15.5,6)})
sns.barplot(data=slaes_state , x= 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states
slaes_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15.5,6)})
sns.barplot(data=slaes_state , x= 'State',y= 'Amount')
```

From above this ghraph we can see that most of the orders & total sales/Amount are from UP, Maharashtra and karnataka respectively

Marital_Status

Occupation


```python
axs = sns.countplot(data= df , x= 'Occupation')
sns.set(rc={'figure.figsize':(20,6)})

for datas in axs.containers:
    axs.bar_label(datas)
```

```python
sales_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data = sales_occupation,x= 'Occupation',y='Amount' )
```

From Above this graphs we can see that most of the buyers are working in IT,aviation and Healthcare sectore


Product Category

```python
PC = sns.countplot(data= df , x= 'Product_Category')
sns.set(rc={'figure.figsize':(20,8)})

for datas in PC.containers:
    PC.bar_label(datas)
```

```python
# sales_Product_Category = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

# sns.set(rc={'figure.figsize':(20,10)})
# sns.barplot(data = sales_Product_Category ,x= 'Product_ID' , y = 'Orders')

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
